```python
# PIE CHART
```

```python
text = {
    'size' : 30,
    'color' : 'Red'
}

colors = ['grey', 'pink', 'green', 'magenta']
```

```python
# WEDGE DENOTES A SLICE IN THE PIE CHART.
wedgeprops = {
    'linewidth' : 3,
    'linestyle' : 'dotted',
    'edgecolor' : 'Red',
}
```

```python
labels = (["goal", "assist", "set pieces", "dribbles"])
explode = [0.4, 0, 0.3, 0.5]
plt.pie([10,40,60,80], radius = 2, startangle = 230, colors = colors, explode = explode, labels = labels, textprops = text, wedgeprops = wedgeprops, rotatelabels = 'True')
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "red", width = 0.5)
plt.show()

plt.barh(x,y, color = 'blue', height = 0.2)
plt.show()
```

```python
values = np.array([40,50,10,20])
plt.subplot(2,2,1)
plt.pie(values)
plt.show()

plt.subplot(2,2,2)
mycolors = ['Red', 'Black', 'Green', 'Yellow']
plt.pie(values, colors = mycolors)
plt.show()

plt.subplot(2,2,3)
mylabels = ['BCT', 'BEX', 'BEL', 'BCA']
plt.pie(values, labels = mylabels)
plt.show()


plt.tight_layout()
```
